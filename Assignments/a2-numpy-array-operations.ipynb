## Genarating Random Integer Numbers with `np.random.randint()`

It is easy to generate random data with numpy. It includes several methods to do it, depending on your data naeeds.

In this case we will focus on generate random integer data with the method `np.random.randint()`, which returns random integers from low (inclusive) to high (exclusive).

The way this method generates the random integers is  from the “discrete uniform” distribution of the specified dtype in the “half-open” interval [low, high). If high is None (the default), then results are from [0, low).

You can read more about *`np.random.randint* [here](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html).

Now all you have to do is to assigne the returned value of `np.random.randint()`, as follows:

```
# This is how you can calculate the mean of an array 
grades = np.random.randint(0, 101, 100)
grades
```

## Ordering the data with `np.sort()`

In statistics the visualization of the data is very important. One can easily see a lot of valuable information just by ordering the data. 

Numpy has a method called `np.sort()`, which returns a sorted copy of an array. For more about how to use this method click [here](https://numpy.org/doc/stable/reference/generated/numpy.sort.html?highlight=sort).

To sort the `grades` in this work, you will just have to pass the array into the method `np.sort()`, as follows:

```
# This shows how to use the method 
sorted_grades = np.sort(grades)
sorted_grades
```

this method has different algortithms implemented inside to order the data. By defualt this method uses the so called "quicksort" algorithm, but you can change the algorithm used with the argument `kind`, where the possible inputs are: `{‘quicksort’, ‘mergesort’, ‘heapsort’, ‘stable’}`

```
np.sort(grades, kind='stable')
```

Also, if you had a bigger array you will be able to specify which axis of the array you would like to sort. In this case the `grades` array is a 1D array, for instance if you change the argument `axis` from its defualt value `1` to `2` you will get an error like the following one:

```
np.sort(grades, axis=2)
```

So now you have it! You sorted the marks of your students in an ascending way, and now we can give it a nice print:

```
print("The sorted grades of the students \n", sorted_grades)
```

## Mean or `np.mean()`

There are several kinds of *mean* in mathematics, but for a data set like the one used in this case, the *mean* is nothing but the *average* of the data set. In other words, the sum of values insede the data set (`grades`) devided by the number of values in the data set (there are `100` values inside `grades`).

You can read more about *mean* [here](https://en.wikipedia.org/wiki/Mean).

Fortunately for us, **NumPy** has its own implementation of mean and you will not have to worry about make the calculation. 

To calculate the *mean* of a data set, you will have to use the method `np.mean()`, which returns the average of the array elements in this case `grades`. The average is taken over the flattened array by default. For more about how to use this method click [here](https://numpy.org/doc/stable/reference/generated/numpy.mean.html?highlight=mean#numpy.mean).

Now all you have to do is to assigne the returned value of `np.mean(grades)`, as follows:

```
# This is how you can calculate the mean of an array 
mean = np.mean(grades)
mean
```

As it was mentioned above, `np.mean()` calculated the sum of all the elements inside the array and devide the sum by the total number of elements inside the array. 

So, what happens inside this method is listed as follows:

```
# This explains how np.mean() works
sum = 0
for i in grades:
  sum += i

mean_explaned = sum / len(grades)
mean_explaned
```

There are several arguments that can be used in `np.mean`. In most cases you will not need to change the defualts, but if you have a bigger array, lets say a matrix, you can change the `axis=` argument and select which axis of the array you want to calculate.

In our case since the `grades` array is a 1D array (`(100,)`) the defualt value for `axis=0` will work, but if you change it, the following will happen:

```
# This illustrates how to break np.mean()
mean = np.mean(grades, axis=1)
```

The reason why it broke is because the array was declared as a 1D horizontal array, for instance `axis=0` will be the horizontal axis. If you change the argument `axis=1`, it will perform the calculation in the second dimension of the array, which in this case does not exist.

So now you have it! You calculated the *mean* mark of your students during the scholar year, and now we can give it a nice print:

```
print("The mean or average mark of the students is: ", mean)
```

## Median or `np.median()`

The *median* is the middle number in a sorted, ascending or descending list of numbers and it can be summarized in 2 escenarios:

* If there is an odd amount of numbers, the median value is the number that is in the middle, with the same amount of numbers below and above.
* If there is an even amount of numbers in the list, the middle pair must be determined, added together, and divided by two to find the median value.

You can read more about the *median* [here](https://www.investopedia.com/terms/m/median.asp#:~:text=Key%20Takeaways-,The%20median%20is%20the%20middle%20number%20in%20a%20sorted%2C%20ascending,the%20average%20of%20the%20values.).

**NumPy** has its own implementation of median.

To calculate the *median* of a data set, you will have to use the method `np.median()`, which compute the median along the specified axis and returns the median of the array elements, in this case `grades`. For more about how to use this method click [here](https://numpy.org/doc/stable/reference/generated/numpy.median.html?highlight=median#numpy.median).

Now all you have to do is to assigne the returned value of `np.median(grades)`, as follows:

```
# This is how you can calculate the mean of an array 
median = np.median(grades)
median
```

As it was mentioned above, `np.median()` calculated the middle element inside the array. 

So, what happens inside this method is that `np.median()` sorted the array `grades` in ascending order and picked the element which was right in the middle, as in the following example.

By using `np.sort()` you can order the grades in ascending order and look ar the output.

There are several arguments that can be used in `np.median()`. In most cases you will not need to change the defualts, but if you have a bigger array, lets say a matrix, you can change the `axis=` argument and select which axis of the array you want to calculate.

In our case since the `grades` array is a 1D array (`(100,)`) the defualt value for `axis=0` will work, but if you change it, the following will happen:

So now you have it! You calculated the *median* mark of your students during the scholar year, and now we can give it a nice print:

## Standard Deviation or `np.std()`

The *standard diviation* is a measure of the amount of variation or dispersion of a set of values, where 

* A low value indicates that the values tend to be close to the mean (or the spected value) of the data set.
* A high value indicates that the values of the data set are spread out over a wider range.

You can read more about the *standard diviation* [here](https://en.wikipedia.org/wiki/Standard_deviation).

**NumPy** has its own implementation of standard diviation.

To calculate the *standard diviation* of a data set, you will have to use the method `np.std()`, which compute the standard deviation and returns the standard deviation, a measure of the spread of a distribution, of the array elements. The standard deviation is computed for the flattened array by default, in this case `grades`. For more about how to use this method click [here](https://numpy.org/doc/stable/reference/generated/numpy.std.html?highlight=numpy%20std#numpy.std).

Now all you have to do is to assigne the returned value of `np.std(grades)`, as follows:

```
# This is how you can calculate the mean of an array 
standard_deviation = np.std(grades)
standard_deviation
```

As it was mentioned before the *standard deviation* measures the variation or dispersion in the data set. 

The next block breaks the steps to calculate it:

```
# This explains how np.std() works

# First we need to calculate the mean or average as before
sum = 0
for i in grades:
  sum += i

mean_explaned = sum / len(grades)

# Then calcualte the numerator of the division 
numerator_up = 0
for i in grades:
  numerator_up += (i - mean_explaned)**2

# Improt math to calculate the square root
import math
# calculate the standard diviation of the data set
math.sqrt(numerator_up / len(grades))
```

There are several arguments that can be used in `np.std()`. In most cases you will not need to change the defualts, but if you have a bigger array, lets say a matrix, you can change the `axis=` argument and select which axis of the array you want to calculate.

In our case since the `grades` array is a 1D array (`(100,)`) the defualt value for `axis=0` will work, but if you change it, the following will happen:

```
# This illustrates how to break np.std()
standard_deviation = np.std(grades, axis=1)
```

The reason why it broke is because the array was declared as a 1D horizontal array, for instance `axis=0` will be the horizontal axis. If you change the argument `axis=1`, it will perform the calculation in the second dimension of the array, which in this case does not exist.

So now you have it! You calculated the *standard deviation* of your students marks during the scholar year, and now we can give it a nice print:

```
print("The standard deviation of the students marks is: ", standard_deviation)
```
